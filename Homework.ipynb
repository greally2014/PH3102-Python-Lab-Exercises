```python
import math                                                     #allows use of cos and sin functions

g=9.81                                                          #states gravity

while True:                                                     #repeats asking for an angle until a valid one is given
    angle=float(input("Enter a launch angle in degrees: "))     #i'm only accepting foward motion, so between 0 and 90 degrees
    if 0<angle<90:                                            
        break                                                   #a valid angle breaks the loop
    else:
        print("Please input an angle between 0 and 90 degrees")     #this will be printed to prompt for a valid answer
        
while True:
    vel=float(input("Enter the velocity of the ball at launch: "))  #repeats asking for a speed until a positive answer is given
    if vel>0:                                                       #this is because we are only accepting foward motion
        break                                                       #i hope you don't mind me adding these parameters
    else:                                                           #just wanted to spice it up a bit
        print("Please input a valid speed: ")

angle=(angle/180)*math.pi                               #converts the angle to radians
time=abs((2*vel*math.sin(angle))/g)                     #uses the inputs prompted to calculate the total time of travel
dist=vel*math.cos(angle)*time                           #uses the inputs prompted to calculate the horizontal distance travelled
dist=round(dist,2)                                      #rounds the distance travelled to two decimal places for convenience

print("The distance travelled is: ",dist," m")          #prints the answer
```

```python
import math
import matplotlib.pyplot as plt

g=9.81
vel=5
angle=(30/180)*math.pi     #converts the angle to radians

time=abs((2*vel*math.sin(angle))/g)  #calculates the time of travel using the two inputs

xdisp=[]                             #array for the x-axis
ydisp=[]                             #array for the y-axis

time = round(time,3)                 #I needed time as an integer for the loop to work
time = time*(10**3)                  #so I round time to 3 decimal places, multiply it by a thousand
time=int(time)                       #now there are no numbers trailing the decimal point

for t in range(0,time,1):            #so now i can convert time to an integer
    t = t/(10**3)                    #as I cycle through time, I divide each step a thousand for the calculations 
    xdisp.append(vel*t*math.cos(angle))                  #adds the calculated xdisp values to array
    ydisp.append((vel*t*math.sin(angle))-(0.5)*g*(t**2)) #adds the calculated ydisp values to array
   
                 
plt.plot(xdisp,ydisp) #plots graph
plt.title("Trajectory of Thrown Ball")
plt.xlabel("Horizontal Displacement (m)")  #labels x-axis
plt.ylabel("Vertical Displacement (m)")    #labels y-axis
plt.show()                                 #displays graph
```
