# Sine Function

#### This is the function we are going to use to generate a 3D Sine wave propagating on the x-axis

$$ f(x,y) = sin \left( \frac{2 * \pi * f * x}{F_s} \right) $$

The variables f and F_s are the frequence and sampling frequency respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm


sigmax, sigmay = 100, 100
nrows, ncols = 1000, 1000

cy, cx = nrows/2, ncols/2
x = np.linspace(0, nrows, nrows)
y = np.linspace(0, ncols, ncols)
X, Y  = np.meshgrid(x, y)
Z = (np.sin((2*np.pi*20*(X))/4))


fig = plt.figure(figsize = (16,10))
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, cmap=cm.gist_rainbow, alpha=0.2, antialiased=False)

ax.set_xlim(0,1000)
ax.set_ylim(0,1000)
ax.set_zlim(-1,1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gaussian Profile')

plt.show()
```
