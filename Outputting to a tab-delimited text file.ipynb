```python
import matplotlib.pyplot as plot
import numpy as np

# Generate our data 
Fs = 5000 # Sampling rate 
f = 50 # the frequency of the signal 

x=np.arange(Fs) 
y=[230*np.cos(2*np.pi*f * (i/Fs)) for i in x] 

plot.figure(figsize=(20,10))
plot.plot(x/Fs, y) 
plot.title('Voltage (V) vs Time (seconds)')
plot.xlabel('Time (s)')
plot.ylabel('Voltage (V)')
plot.show 

np.savetxt("VvsTDanielGreally.txt",np.transpose([x/Fs, y ]),fmt=("%.5f","%.5f"),delimiter="\t", newline="\n", header="Time(s)   Voltage(V)")
```
