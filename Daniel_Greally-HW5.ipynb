```python
import matplotlib.pyplot as plot    #imports what we need for the calculations and graphing
import math
import numpy as np

with open('StarData.txt') as StarData:         #finds and opens the text file with the data we need for calculations
    lines = StarData.read().splitlines()       #i believe this creates columns, not completely sure


#Set up our data arrays
StarID = []
V_Mag = []
B_V = []
Actual_Mag = []
Parallaxes = []       # Units are in arcsec  
Parallax_Error = []   # Units are in milliarcsec  

for line in lines:
    Columns = line.split()         #creates column entries
    StarID.append(Columns[0])      #adds the values each loop gives to a particular column
    V_Mag.append(Columns[1])
    B_V.append(Columns[2])
    Parallaxes.append(Columns[3])
    Parallax_Error.append(Columns[4])
    d = 1/float(Columns[3])   #calculates d from p
    Actual_Mag.append(float(Columns[1])- 5*math.log(d/10))  #formula for absolute magnitude

    
# Plot our data - 
plot.figure(figsize=(20,10))                            #gives graph dimensions
plot.scatter(B_V, Actual_Mag, marker='*', c='b', s=0.8)
plot.xlabel("B-V Colour")                               #titles the graph x axis
plot.ylabel("Absolute Magnitude $\mathregular{m_v}$")   #titles the graph y axis
plot.title("Absolute Magnitude $\mathregular{m_v}$ vs B-V Colour") #titles graph
plot.show() #displays graph

```
